## **Cek Outliers**

```python
results = []
cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

**Insight:**

Variabel Impressions tidak memiliki outliers (0%) yang menunjukkan distribusi data tayangan lebih stabil dan tidak terdapat nilai ekstrem.

Pada variabel Clicks, hanya terdapat 0,54% outliers. Presentase ini sangat kecil sehingga kurang berpengaruh dengan analisisnya.

Variabel Conversions memiliki presentase outliers tertinggi yaitu 4,36%. Ini menunjukkan ada beberapa kampanye yang performanya berbeda dari yang lain.

# **EDA (Eksplorasi Data)**

## **1. Comparison (Perbandingan)**

```python
marketing_campaign = df.groupby('Channel')['Conversions'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=marketing_campaign.index, y=marketing_campaign.values, palette='viridis', hue=marketing_campaign.index, legend=False)
plt.title('Marketing Campaign Data')
plt.xlabel('Channel')
plt.ylabel('Conversions')
plt.xticks(rotation=45)
plt.show()
```

**Insight:**

Berdasarkan bar chart, channel dengan nilai konversi tertinggi menunjukkan pada channel Facebook. Channel dengan Conversions tertinggi menghasilkan aksi yang paling efektif seperti melakukan pembelian atau pendaftaran kampanye. Channel dengan total konversi tertinggi juga platform yang optimal dalam strategi pemasaran, seperti kualitas audiens, kontennya, dan efektivitas kepada targetnya.

Tetapi diperlukan analisis perbandingan antara Impressions dan Conversions untuk mengetahui efektivitas masing-masing channel.

## **2. Composition (Komposisi)**

```python
marketing_campaign = df.groupby('Channel')['Impressions'].sum().sort_values(ascending=False)
marketing_campaign_top3 = marketing_campaign.head(3)

plt.figure(figsize=(10, 8))
marketing_campaign_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Channel Teratas Berdasarkan Total Impressions')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Insight:**

Berdasarkan pie chart, tiga channel dengan total Impressions tertinggi yang pertama adalah Tiktok, kedua adalah Facebook, dan yang ketiga adalah Email. Tiktok menjadi channel dengan kontribusi impressions terbesar, yaitu sebesar 34,3%, lalu Facebook sebesar 34,1%, dan Email sebesar 31,6%.

Tetapi tinggi Impressions belum tentu mengefek pada Conversions. Diperlukan analisis lanjutan untuk mengetahui apakah Channel dengan Impressions tinggi menghasilkan Conversions yang tinggi juga.

## **3. Distribution (Distribusi)**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Impressions'], bins=20, kde=True)
plt.title('Distribusi Frekuensi Impressions')
plt.xlabel('Impressions')
plt.ylabel('Frekuensi')
plt.show()
```

Tidak ada kemiringan ekstrem ke kanan maupun ke kiri dan frekuensi pada tiap rentang terlihat stabil. ## **4. Relationship (Hubungan)**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Impressions', 'Conversions']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Impressions dan Conversions')
plt.show()
```

**Insight:**

Berdasarkan heatmap, hasil korelasi menunjukkan bahwa nilai korelasi sebesar 0.69 antara Impressions dan Conversions. Nilai ini menunjukkan hubungan positif yang kuat antara jumlah tayangan iklan dan jumlah konversi. Semakin tinggi impressions yang didapat kampanye, semakin besar kemungkinan terjadi konversi. Karena korelasinya tidak mendekati 1, terdapat faktor lain selain impressions yang memengaruhi conversions, yaitu kualitas konten, target audiens, atau strategi pemasaran yang digunakan.
